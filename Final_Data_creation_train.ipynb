# for df_family

```python
# merged_train_trans=pd.merge(train,transactions,on=['date','store_nbr'])
merged_train_oil=pd.merge(train,oil, on=['date'])
df_family= merged_train_oil
df_family.rename(columns={'dcoilwtico':'oil_prices'},inplace=True)
df_family.drop(['id','store_nbr'],axis=1,inplace=True)

family_column=df_family.pop('family')
df_family.insert(4,'family',family_column)
sales_column=df_family.pop('sales')
df_family.insert(4,'sales',sales_column)
# df_family.to_csv('df_family.csv',index=False)
df_family.head(10)
```

# for df_trans

```python
train_nofamily=train
train_nofamily.drop(['family'],axis=1,inplace=True)
merged_train_nofamily_oil=pd.merge(train_nofamily,oil, on=['date'])
df_trans=pd.merge(merged_train_nofamily_oil,transactions,on=['date','store_nbr'])

df_trans.rename(columns={'dcoilwtico':'oil_prices'},inplace=True)
df_trans.drop(['id','store_nbr'],axis=1,inplace=True)
sales_column=df_trans.pop('sales')
df_trans.insert(4,'sales',sales_column)
# df_trans.to_csv('df_trans.csv',index=False)
df_trans.head()
```

```python
import pandas as pd

# Assuming your DataFrame is named df
# Group by 'date' and aggregate 'sales', 'transactions', and 'onpromotion'
grouped_df = df_trans.groupby('date').agg({'sales': 'sum', 'transactions': 'first', 'onpromotion': 'sum','oil_prices': 'first'}).reset_index()

# Display the resulting DataFrame
grouped_df
```

```python
trans_col_=grouped_df.pop('transactions')
grouped_df.insert(4,'transactions',trans_col_)

sales_col_=grouped_df.pop('sales')
grouped_df.insert(4,'sales',sales_col_)

grouped_df.to_csv('new_trans_data.csv',index=False)
grouped_df.head()
```
